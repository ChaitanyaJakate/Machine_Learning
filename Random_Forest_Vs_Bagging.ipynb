```python
import numpy as np
import pandas as pd
from sklearn.datasets import make_classification
```

```python
X,y=make_classification(n_redundant=0,n_features=5,n_informative=5,n_clusters_per_class=1)
```

```python
X.shape
```

```python
y.shape
```

```python
df=pd.DataFrame(X,columns=['col1','col2','col3','col4','col5'])
df['target']=y
df.head()
```

```python
import matplotlib.pyplot as plt
```

```python
#plt.scatter(X,y)
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
from sklearn.ensemble import BaggingClassifier
bag=BaggingClassifier(max_features=2)
```

```python
bag.fit(df.iloc[:,0:5],df.iloc[:,-1])
```

```python
from sklearn.tree import plot_tree
```

```python
plt.figure(figsize=(12,12))
plot_tree(bag.estimators_[0])
plt.show()
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rfc=RandomForestClassifier(max_features=2)
```

```python
rfc.fit(df.iloc[:,0:5],df.iloc[:,-1])
```

```python
plt.figure(figsize=(12,12))
plot_tree(rfc.estimators_[0])
plt.show()
```
